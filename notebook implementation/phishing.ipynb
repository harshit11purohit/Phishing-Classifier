```python
import matplotlib.pyplot as plot
import seaborn as sns

data_sample = data.sample(n=min(1000, len(data)), random_state=42)

plot.figure(figsize=(15, 20), facecolor='white')
plotnumber = 1

# Loop through columns (excluding the target 'Result')
for column in data_sample.drop(['Result'], axis=1):
    if plotnumber <= 36:  # Limits to a 12x3 grid
        ax = plot.subplot(12, 3, plotnumber)
        
        # Use the sampled data for the plot
        sns.countplot(x=data_sample[column], palette='viridis')
        
        plot.xlabel(column, fontsize=10)
        plot.ylabel('Count', fontsize=8)
        plotnumber += 1

plot.tight_layout() # Prevents labels from overlapping
plot.show()
```
