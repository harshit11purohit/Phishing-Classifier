```python
import os
import pandas as pd
import pymongo

def upload_files_to_mongodb(mongo_client_con_string, database_name, datasets_dir_name):
    """
    Finds all CSV files in a directory and uploads them to MongoDB Atlas.
    """
    try:
        # Initialize connection
        client = pymongo.MongoClient(mongo_client_con_string)
        db = client[database_name]
        
        # Look for CSVs in the folder
        for file in os.listdir(datasets_dir_name):
            if file.endswith('.csv'):
                collection_name = file.split('.')[0]
                collection = db[collection_name]
                
                # Check if data already exists to avoid duplicates
                if collection.count_documents({}) > 0:
                    print(f"⏩ {collection_name} already has data. Skipping to avoid duplicates.")
                    continue

                file_path = os.path.join(datasets_dir_name, file)
                df = pd.read_csv(file_path)
                
                # Upload
                collection.insert_many(df.to_dict("records"))
                print(f"✅ Successfully uploaded {len(df)} rows from '{file}' to Atlas.")
                
    except Exception as e:
        print(f"❌ An error occurred: {e}")

# --- EXECUTION ---
# Use your verified details
CONNECTION_STRING = "mongodb+srv://harshit_purohit:hp123456@cluster0.ijjud8u.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0"
DB_NAME = "phising"
DATA_DIR = "."  # Since your CSV is in the same folder as the notebook

upload_files_to_mongodb(CONNECTION_STRING, DB_NAME, DATA_DIR)
```
